```python
import tensorflow as tf
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = tf.placeholder(tf.float32, shape=(None, 4))
```

```python
y = tf.placeholder(tf.float32, shape=(None, 1))
```

```python
is_training = tf.placeholder(tf.bool)
```

```python
w = tf.layers.dense(x, 10, activation=tf.nn.relu)
```

```python
w1 = tf.layers.batch_normalization(w, training=is_training)
```

```python
output = tf.layers.dense(w1, 1)
```

```python
loss = tf.reduce_mean((output - y) ** 2)
```

```python
update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
```

```python
with tf.control_dependencies(update_ops):
    train_op = tf.train.RMSPropOptimizer(0.001).minimize(loss)
```

```python
w_ = 1 + np.random.random((4, 1))
```

```python
sess = tf.InteractiveSession()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
train_losses = []
val_losses = []
bias = 5
for i in range(5000):
    x_ = bias + np.random.random((100, 4))
    y_ = np.dot(x_, w_)
    _, loss_ = sess.run([train_op, loss], 
                        {x: x_, y:y_, is_training: True})
    train_losses.append(loss_)
    if i % 100 == 0:
        x_ = bias + np.random.random((100, 4))
        y_ = np.dot(x_, w_)
        val_loss = sess.run(loss, {x: x_, y:y_, is_training: False})
        val_losses.append(val_loss)
```

```python
plt.plot(train_losses[-10:])
```

```python
plt.plot(val_losses[-10:])
```

```python
gs = tf.placeholder(tf.int32)
```

```python
lr = tf.train.cosine_decay_restarts(0.001, gs, 1000)
```

```python
sess = tf.InteractiveSession()
```

```python
ts = []
for i in range(10000):
    t = sess.run(lr, {gs: i})
    ts.append(t)
```
